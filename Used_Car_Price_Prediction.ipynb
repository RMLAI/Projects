```
import pandas as pd
url = 'https://raw.githubusercontent.com/krishnaik06/Car-Price-Prediction/master/car%20data.csv'
df= pd.read_csv(url)
df.head()
```

```
df.groupby('Car_Name').mean()
```

```
#EDA
df.describe()
```

```
#checking the cars count by years
df['Year'].value_counts().plot(kind='bar');
```

```
#checking the count of cars by brand
df['Car_Name'].value_counts().plot(kind='bar',figsize=(20,20))
                                  
```

```
#using pivot tables to explore the data
table1=pd.pivot_table(df,values='Selling_Price',index='Car_Name',columns='Fuel_Type',aggfunc='mean',margins='True',margins_name='average')
table1
```

```
#changing the default number of visible rows
pd.set_option('max_rows', 99999)
```

```
#using pivot tables to explore the data
table2=pd.pivot_table(df,values='Selling_Price',index=['Car_Name','Year'],columns='Fuel_Type',aggfunc='mean',margins='True',margins_name='average')
table2
```

```
#creating a duplicate dataframe
df_temp=df
```

```
#creating a new column called age by subtracting present year from model year
df_temp['Current_Year']=2020
df_temp['Vehicle_Age']=df_temp['Current_Year']-df_temp['Year']
df_temp.head()
```

```
#dropping Age, Current_Year variables
df_temp.drop(['Current_Year','Age'],axis=1,inplace=True)
df_temp.head()
```
